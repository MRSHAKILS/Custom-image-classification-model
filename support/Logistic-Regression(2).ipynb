```python

import os
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score 
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.metrics import (classification_report, confusion_matrix, accuracy_score, 
                           precision_recall_fscore_support)
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif, VarianceThreshold
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
from skimage import feature, filters, measure
from scipy import ndimage
import itertools
from tensorflow.keras.applications import VGG16, ResNet50, InceptionV3
from tensorflow.keras.applications.vgg16 import preprocess_input as vgg16_preprocess
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet50_preprocess
from tensorflow.keras.applications.inception_v3 import preprocess_input as inception_preprocess
import tensorflow as tf

warnings.filterwarnings('ignore')
plt.style.use('default')
np.random.seed(42)
tf.random.set_seed(42)

print("ğŸš€ ENHANCED LOGISTIC REGRESSION + CNN FEATURES")
print("="*60)
print("ğŸ§  CNN Feature Extraction (VGG16 + ResNet50 + InceptionV3)")
print("ğŸ“Š Advanced Logistic Regression with Hyperparameter Tuning")
print("ğŸ¯ Target: >90% Accuracy")
print("="*60)

# =============================================================================
# CELL 1: Enhanced Data Loading with Augmentation
# =============================================================================

def apply_data_augmentation(image):
    """Apply comprehensive data augmentation"""
    augmented_images = [image]  # Original image
    
    # Rotation variations
    for angle in [15, -15, 30, -30, 45, -45]:
        center = (image.shape[1]//2, image.shape[0]//2)
        matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
        rotated = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))
        augmented_images.append(rotated)
    
    # Flipping
    flipped_h = cv2.flip(image, 1)  # Horizontal flip
    flipped_v = cv2.flip(image, 0)  # Vertical flip
    flipped_both = cv2.flip(image, -1)  # Both axes
    augmented_images.extend([flipped_h, flipped_v, flipped_both])
    
    # Brightness and contrast adjustments
    bright_high = cv2.convertScaleAbs(image, alpha=1.3, beta=30)
    bright_low = cv2.convertScaleAbs(image, alpha=0.7, beta=-30)
    contrast_high = cv2.convertScaleAbs(image, alpha=1.5, beta=0)
    contrast_low = cv2.convertScaleAbs(image, alpha=0.5, beta=0)
    augmented_images.extend([bright_high, bright_low, contrast_high, contrast_low])
    
    # Noise and blur
    blurred = cv2.GaussianBlur(image, (5, 5), 0)
    blurred_strong = cv2.GaussianBlur(image, (7, 7), 0)
    
    # Add slight noise
    noise = np.random.normal(0, 0.05, image.shape)
    noisy = np.clip(image + noise, 0, 1)
    
    augmented_images.extend([blurred, blurred_strong, noisy])
    
    # Scaling variations
    h, w = image.shape[:2]
    # Zoom in (crop center and resize)
    crop_size = int(min(h, w) * 0.8)
    start_h, start_w = (h - crop_size) // 2, (w - crop_size) // 2
    cropped = image[start_h:start_h+crop_size, start_w:start_w+crop_size]
    zoomed = cv2.resize(cropped, (w, h))
    augmented_images.append(zoomed)
    
    return augmented_images

def load_enhanced_dataset(dataset_path='/kaggle/input/animalclassification/datset_new', 
                         target_size=(224, 224), apply_augmentation=True):
    """Load dataset optimized for CNN feature extraction"""
    
    print("\nğŸ“ ENHANCED DATASET LOADING FOR CNN")
    print("-" * 40)
    
    classes = ['dog', 'cow', 'cat', 'lamb', 'zebra']
    images = []
    labels = []
    
    dataset_path = Path(dataset_path)
    
    print(f"Dataset path: {dataset_path}")
    print(f"Target size: {target_size} (optimized for CNN)")
    print(f"Data augmentation: {'ON' if apply_augmentation else 'OFF'}")
    
    if not dataset_path.exists():
        print(f"âŒ Dataset not found at {dataset_path}")
        return None, None
    
    total_original = 0
    total_augmented = 0
    
    for class_name in classes:
        class_path = dataset_path / class_name
        
        if not class_path.exists():
            print(f"âš ï¸  Class '{class_name}' not found")
            continue
            
        # Get image files
        image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff']
        image_files = []
        for ext in image_extensions:
            image_files.extend(class_path.glob(f'*{ext}'))
            image_files.extend(class_path.glob(f'*{ext.upper()}'))
        
        original_count = len(image_files)
        total_original += original_count
        
        print(f"ğŸ“¸ {class_name.capitalize()}: {original_count} original images")
        
        # Load and process images
        class_images = []
        for img_path in image_files:
            try:
                img = cv2.imread(str(img_path))
                if img is not None:
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    img = cv2.resize(img, target_size)
                    
                    if apply_augmentation:
                        # Apply comprehensive augmentation
                        augmented = apply_data_augmentation(img)
                        for aug_img in augmented:
                            # Normalize to [0,1]
                            norm_img = aug_img.astype(np.float32) / 255.0
                            class_images.append(norm_img)
                    else:
                        norm_img = img.astype(np.float32) / 255.0
                        class_images.append(norm_img)
                        
            except Exception as e:
                print(f"âš ï¸  Error loading {img_path.name}: {e}")
                continue
        
        # Add to main dataset
        images.extend(class_images)
        labels.extend([class_name] * len(class_images))
        
        augmented_count = len(class_images)
        total_augmented += augmented_count
        print(f"   âœ… Total after augmentation: {augmented_count} images")
    
    print(f"\nğŸ“Š Dataset Summary:")
    print(f"   Original images: {total_original}")
    print(f"   After augmentation: {total_augmented}")
    print(f"   Augmentation factor: {total_augmented/total_original:.1f}x")
    
    if len(images) == 0:
        return None, None
    
    X = np.array(images)
    y = np.array(labels)
    
    print(f"ğŸ“ Final shape: {X.shape}")
    return X, y

# Load enhanced dataset
X_raw, y_raw = load_enhanced_dataset(apply_augmentation=True)

if X_raw is None:
    print("âŒ Failed to load dataset")
else:
    print(f"âœ… Dataset loaded: {len(X_raw)} samples")

# =============================================================================
# CELL 2: CNN Feature Extraction
# =============================================================================

def extract_cnn_features(images, batch_size=32):
    """Extract features using multiple pre-trained CNN models"""
    
    print("\nğŸ§  CNN FEATURE EXTRACTION")
    print("-" * 35)
    
    # Load pre-trained models
    print("ğŸ“¥ Loading pre-trained CNN models...")
    
    # VGG16
    vgg16 = VGG16(weights='imagenet', include_top=False, pooling='avg', input_shape=(224, 224, 3))
    vgg16.trainable = False
    
    # ResNet50
    resnet50 = ResNet50(weights='imagenet', include_top=False, pooling='avg', input_shape=(224, 224, 3))
    resnet50.trainable = False
    
    # InceptionV3
    inception = InceptionV3(weights='imagenet', include_top=False, pooling='avg', input_shape=(224, 224, 3))
    inception.trainable = False
    
    print("âœ… CNN models loaded successfully")
    
    # Extract features in batches
    num_samples = len(images)
    num_batches = (num_samples + batch_size - 1) // batch_size
    
    vgg16_features = []
    resnet50_features = []
    inception_features = []
    
    print(f"ğŸ”„ Extracting features from {num_samples} images in {num_batches} batches...")
    
    for i in range(num_batches):
        start_idx = i * batch_size
        end_idx = min((i + 1) * batch_size, num_samples)
        batch_images = images[start_idx:end_idx]
        
        if i % 10 == 0:
            print(f"   Processing batch {i+1}/{num_batches}...")
        
        # Preprocess for each model
        batch_images_255 = (batch_images * 255.0).astype(np.uint8)
        
        # VGG16 features
        vgg16_input = vgg16_preprocess(batch_images_255.copy())
        vgg16_batch_features = vgg16.predict(vgg16_input, verbose=0)
        vgg16_features.append(vgg16_batch_features)
        
        # ResNet50 features
        resnet50_input = resnet50_preprocess(batch_images_255.copy())
        resnet50_batch_features = resnet50.predict(resnet50_input, verbose=0)
        resnet50_features.append(resnet50_batch_features)
        
        # InceptionV3 features
        inception_input = inception_preprocess(batch_images_255.copy())
        inception_batch_features = inception.predict(inception_input, verbose=0)
        inception_features.append(inception_batch_features)
    
    # Concatenate all batches
    vgg16_features = np.vstack(vgg16_features)
    resnet50_features = np.vstack(resnet50_features)
    inception_features = np.vstack(inception_features)
    
    print(f"âœ… Feature extraction completed:")
    print(f"   VGG16 features: {vgg16_features.shape}")
    print(f"   ResNet50 features: {resnet50_features.shape}")
    print(f"   InceptionV3 features: {inception_features.shape}")
    
    # Combine all CNN features
    combined_features = np.concatenate([vgg16_features, resnet50_features, inception_features], axis=1)
    
    print(f"ğŸ”— Combined CNN features: {combined_features.shape}")
    
    return combined_features

def extract_traditional_features(images):
    """Extract traditional computer vision features"""
    
    print("\nğŸ¨ TRADITIONAL FEATURE EXTRACTION")
    print("-" * 40)
    
    features_list = []
    
    for i, img in enumerate(images):
        if i % 1000 == 0:
            print(f"Processing image {i+1}/{len(images)}...")
        
        feature_vector = []
        
        # Convert to different formats
        img_uint8 = (img * 255).astype(np.uint8)
        gray = cv2.cvtColor(img_uint8, cv2.COLOR_RGB2GRAY)
        hsv = cv2.cvtColor(img_uint8, cv2.COLOR_RGB2HSV)
        lab = cv2.cvtColor(img_uint8, cv2.COLOR_RGB2LAB)
        
        # 1. Enhanced Color Statistics
        for channel in range(3):
            # RGB statistics
            feature_vector.extend([
                np.mean(img[:, :, channel]),
                np.std(img[:, :, channel]),
                np.median(img[:, :, channel]),
                np.percentile(img[:, :, channel], 25),
                np.percentile(img[:, :, channel], 75),
                np.min(img[:, :, channel]),
                np.max(img[:, :, channel])
            ])
            
            # HSV statistics
            feature_vector.extend([
                np.mean(hsv[:, :, channel]),
                np.std(hsv[:, :, channel]),
                np.median(hsv[:, :, channel])
            ])
            
            # LAB statistics
            feature_vector.extend([
                np.mean(lab[:, :, channel]),
                np.std(lab[:, :, channel])
            ])
        
        # 2. Advanced Texture Features
        try:
            # LBP with multiple radii
            for radius in [1, 2, 3]:
                P = 8 * radius
                lbp = feature.local_binary_pattern(gray, P=P, R=radius, method='uniform')
                lbp_hist, _ = np.histogram(lbp.ravel(), bins=P+2, range=(0, P+2))
                lbp_hist = lbp_hist.astype(float)
                lbp_hist /= (lbp_hist.sum() + 1e-7)
                feature_vector.extend(lbp_hist[:10])  # Take first 10 bins
        except:
            feature_vector.extend([0] * 30)
        
        # 3. Enhanced Edge Features
        try:
            # Multiple edge detection methods
            edges_canny = cv2.Canny(gray, 50, 150)
            edges_sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
            edges_sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
            edges_laplacian = cv2.Laplacian(gray, cv2.CV_64F)
            
            feature_vector.extend([
                np.sum(edges_canny > 0) / edges_canny.size,
                np.mean(np.abs(edges_sobel_x)),
                np.std(edges_sobel_x),
                np.mean(np.abs(edges_sobel_y)),
                np.std(edges_sobel_y),
                np.mean(np.abs(edges_laplacian)),
                np.std(edges_laplacian)
            ])
        except:
            feature_vector.extend([0] * 7)
        
        # 4. Shape and Geometric Features
        try:
            # Moments and contours
            moments = cv2.moments(gray)
            if moments['m00'] != 0:
                hu_moments = cv2.HuMoments(moments).flatten()
                feature_vector.extend(hu_moments)
                
                # Centroid
                cx = moments['m10'] / moments['m00']
                cy = moments['m01'] / moments['m00']
                feature_vector.extend([cx/gray.shape[1], cy/gray.shape[0]])
            else:
                feature_vector.extend([0] * 9)
        except:
            feature_vector.extend([0] * 9)
        
        # 5. Enhanced Histogram Features
        for channel in range(3):
            hist, _ = np.histogram(img[:, :, channel], bins=32, range=(0, 1))
            hist = hist.astype(float)
            hist /= (hist.sum() + 1e-7)
            feature_vector.extend(hist)
        
        # 6. Texture Energy and Entropy
        try:
            # Gray level co-occurrence matrix features
            glcm = feature.greycomatrix(gray, [1], [0], symmetric=True, normed=True)
            contrast = feature.greycoprops(glcm, 'contrast')[0, 0]
            dissimilarity = feature.greycoprops(glcm, 'dissimilarity')[0, 0]
            homogeneity = feature.greycoprops(glcm, 'homogeneity')[0, 0]
            energy = feature.greycoprops(glcm, 'energy')[0, 0]
            
            feature_vector.extend([contrast, dissimilarity, homogeneity, energy])
        except:
            feature_vector.extend([0, 0, 0, 0])
        
        features_list.append(feature_vector)
    
    features_array = np.array(features_list)
    print(f"âœ… Traditional features shape: {features_array.shape}")
    
    return features_array

# Extract CNN features
if X_raw is not None:
    print("ğŸ§  Extracting CNN features...")
    cnn_features = extract_cnn_features(X_raw)
    
    print("ğŸ¨ Extracting traditional features...")
    traditional_features = extract_traditional_features(X_raw)
    
    # Combine all features
    X_combined = np.concatenate([cnn_features, traditional_features], axis=1)
    print(f"ğŸ”— Combined features shape: {X_combined.shape}")

# =============================================================================
# CELL 3: Enhanced Data Preprocessing
# =============================================================================

def create_stratified_splits(X, y, test_size=0.2, val_size=0.2):
    """Create stratified splits with proper encoding"""
    
    print(f"\nğŸ“Š CREATING STRATIFIED SPLITS")
    print("-" * 35)
    
    # Encode labels
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)
    
    # First split: train+val vs test
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y_encoded, test_size=test_size, random_state=42, stratify=y_encoded
    )
    
    # Second split: train vs val
    val_size_adj = val_size / (1 - test_size)
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=val_size_adj, random_state=42, stratify=y_temp
    )
    
    print(f"ğŸ“ˆ Split distribution:")
    print(f"   Train: {len(X_train):,} samples ({len(X_train)/len(X)*100:.1f}%)")
    print(f"   Val:   {len(X_val):,} samples ({len(X_val)/len(X)*100:.1f}%)")
    print(f"   Test:  {len(X_test):,} samples ({len(X_test)/len(X)*100:.1f}%)")
    
    # Check class distribution
    print(f"\nğŸ“Š Class distribution:")
    for i, class_name in enumerate(label_encoder.classes_):
        train_count = np.sum(y_train == i)
        val_count = np.sum(y_val == i)
        test_count = np.sum(y_test == i)
        print(f"   {class_name}: Train={train_count}, Val={val_count}, Test={test_count}")
    
    return X_train, X_val, X_test, y_train, y_val, y_test, label_encoder

def apply_advanced_preprocessing(X_train, X_val, X_test, variance_threshold=0.01, n_components=1000):
    """Apply comprehensive preprocessing pipeline"""
    
    print(f"\nâš™ï¸  ADVANCED PREPROCESSING PIPELINE")
    print("-" * 40)
    
    # 1. Remove low-variance features
    print("ğŸ” Removing low-variance features...")
    variance_selector = VarianceThreshold(threshold=variance_threshold)
    X_train_var = variance_selector.fit_transform(X_train)
    X_val_var = variance_selector.transform(X_val)
    X_test_var = variance_selector.transform(X_test)
    print(f"   {X_train.shape[1]} â†’ {X_train_var.shape[1]} features")
    
    # 2. Feature scaling
    print("ğŸ“ Applying StandardScaler...")
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_var)
    X_val_scaled = scaler.transform(X_val_var)
    X_test_scaled = scaler.transform(X_test_var)
    print("   âœ… Features standardized")
    
    # 3. Feature selection using statistical tests
    print("ğŸ¯ Selecting best features...")
    k_best = min(2000, X_train_scaled.shape[1])  # Select top features
    selector = SelectKBest(score_func=f_classif, k=k_best)
    X_train_selected = selector.fit_transform(X_train_scaled, y_train)
    X_val_selected = selector.transform(X_val_scaled)
    X_test_selected = selector.transform(X_test_scaled)
    print(f"   {X_train_scaled.shape[1]} â†’ {X_train_selected.shape[1]} features")
    
    # 4. PCA for dimensionality reduction
    print("ğŸ”„ Applying PCA...")
    n_components = min(n_components, X_train_selected.shape[1], X_train_selected.shape[0] - 1)
    pca = PCA(n_components=n_components, random_state=42)
    X_train_pca = pca.fit_transform(X_train_selected)
    X_val_pca = pca.transform(X_val_selected)
    X_test_pca = pca.transform(X_test_selected)
    
    explained_variance = np.sum(pca.explained_variance_ratio_)
    print(f"   {X_train_selected.shape[1]} â†’ {n_components} components")
    print(f"   Explained variance: {explained_variance:.4f} ({explained_variance*100:.2f}%)")
    
    print(f"\nâœ… Preprocessing completed!")
    print(f"   Final feature shape: {X_train_pca.shape}")
    
    return (X_train_pca, X_val_pca, X_test_pca, 
            variance_selector, scaler, selector, pca)

# Create splits and preprocess
if X_raw is not None:
    print("ğŸ“Š Creating data splits...")
    X_train, X_val, X_test, y_train, y_val, y_test, label_encoder = create_stratified_splits(
        X_combined, y_raw
    )
    
    print("âš™ï¸  Preprocessing features...")
    X_train_final, X_val_final, X_test_final, var_selector, scaler, feature_selector, pca = apply_advanced_preprocessing(
        X_train, X_val, X_test
    )

# =============================================================================
# CELL 4: Enhanced Logistic Regression with Hyperparameter Tuning
# =============================================================================

def train_enhanced_logistic_regression(X_train, y_train, X_val, y_val):
    """Train logistic regression with efficient hyperparameter tuning"""
    
    print(f"\nğŸ“Š ENHANCED LOGISTIC REGRESSION TRAINING")
    print("-" * 50)
    
    # Stage 1: Quick screening with reduced parameters
    print("ğŸš€ Stage 1: Quick parameter screening...")
    
    quick_param_grid = {
        'C': [0.01, 0.1, 1.0, 10.0, 100.0],
        'solver': ['lbfgs', 'liblinear', 'saga'],
        'penalty': ['l1', 'l2', None],
        'max_iter': [1000]
    }
    
    # Use GridSearchCV for efficient parallel processing
    lr_base = LogisticRegression(random_state=42, multi_class='ovr')
    grid_search = GridSearchCV(
        lr_base, 
        quick_param_grid, 
        cv=3,  # Reduced CV for speed
        scoring='accuracy',
        n_jobs=-1,  # Use all cores
        verbose=1
    )
    
    print("   Running GridSearchCV (using all CPU cores)...")
    grid_search.fit(X_train, y_train)
    
    print(f"âœ… Stage 1 completed!")
    print(f"   Best CV score: {grid_search.best_score_:.6f}")
    print(f"   Best params: {grid_search.best_params_}")
    
    # Stage 2: Fine-tune around best parameters
    print(f"\nğŸ¯ Stage 2: Fine-tuning around best parameters...")
    
    best_C = grid_search.best_params_['C']
    best_solver = grid_search.best_params_['solver']
    best_penalty = grid_search.best_params_['penalty']
    
    # Create fine-tuning grid around best values
    if best_C <= 0.1:
        C_fine = [best_C/5, best_C/2, best_C, best_C*2, best_C*5]
    elif best_C >= 10:
        C_fine = [best_C/5, best_C/2, best_C, best_C*2, best_C*5]
    else:
        C_fine = [best_C/3, best_C/1.5, best_C, best_C*1.5, best_C*3]
    
    fine_param_grid = {
        'C': C_fine,
        'solver': [best_solver],
        'penalty': [best_penalty],
        'max_iter': [1000, 2000, 3000]
    }
    
    # Handle solver-penalty compatibility for fine-tuning
    if best_solver == 'liblinear' and best_penalty not in ['l1', 'l2']:
        fine_param_grid['penalty'] = ['l2']
    elif best_solver in ['lbfgs', 'newton-cg'] and best_penalty not in ['l2', None]:
        fine_param_grid['penalty'] = ['l2']
    
    fine_grid_search = GridSearchCV(
        lr_base,
        fine_param_grid,
        cv=5,  # Full CV for final model
        scoring='accuracy',
        n_jobs=-1,
        verbose=1
    )
    
    print("   Fine-tuning parameters...")
    fine_grid_search.fit(X_train, y_train)
    
    best_model = fine_grid_search.best_estimator_
    best_params = fine_grid_search.best_params_
    
    # Evaluate on validation set
    val_score = best_model.score(X_val, y_val)
    
    print(f"\nğŸ† BEST LOGISTIC REGRESSION MODEL FOUND!")
    print(f"âœ… Best CV score: {fine_grid_search.best_score_:.6f}")
    print(f"âœ… Validation accuracy: {val_score:.6f} ({val_score*100:.4f}%)")
    print(f"ğŸ“‹ Final best parameters:")
    for param, value in best_params.items():
        print(f"   {param}: {value}")
    
    # Stage 3: Test alternative configurations if performance is low
    if val_score < 0.85:  # If not meeting expectations
        print(f"\nğŸ”„ Stage 3: Testing alternative configurations...")
        
        alternative_configs = [
            {'C': 0.1, 'solver': 'saga', 'penalty': 'elasticnet', 'l1_ratio': 0.5, 'max_iter': 2000},
            {'C': 1.0, 'solver': 'saga', 'penalty': 'elasticnet', 'l1_ratio': 0.3, 'max_iter': 2000},
            {'C': 10.0, 'solver': 'saga', 'penalty': 'elasticnet', 'l1_ratio': 0.7, 'max_iter': 2000},
            {'C': 0.01, 'solver': 'liblinear', 'penalty': 'l1', 'max_iter': 3000},
            {'C': 100.0, 'solver': 'lbfgs', 'penalty': 'l2', 'max_iter': 3000},
        ]
        
        for i, config in enumerate(alternative_configs):
            try:
                print(f"   Testing alternative config {i+1}/5...")
                alt_model = LogisticRegression(random_state=42, multi_class='ovr', **config)
                alt_model.fit(X_train, y_train)
                alt_score = alt_model.score(X_val, y_val)
                
                if alt_score > val_score:
                    print(f"   âœ… Better config found! Score: {alt_score:.6f}")
                    best_model = alt_model
                    best_params = config
                    val_score = alt_score
                    
            except Exception as e:
                print(f"   âš ï¸  Config {i+1} failed: {str(e)[:50]}...")
                continue
    
    # Final cross-validation
    print(f"\nğŸ”„ Final cross-validation...")
    cv_scores = cross_val_score(best_model, X_train, y_train, cv=5, scoring='accuracy', n_jobs=-1)
    print(f"ğŸ“Š CV Scores: {cv_scores}")
    print(f"ğŸ“ˆ CV Mean: {cv_scores.mean():.6f} Â± {cv_scores.std():.6f}")
    
    return best_model, best_params, val_score

# Train enhanced logistic regression
if X_raw is not None:
    print("ğŸ“Š Training enhanced logistic regression...")
    best_lr_model, best_params, validation_score = train_enhanced_logistic_regression(
        X_train_final, y_train, X_val_final, y_val
    )

# =============================================================================
# CELL 5: Final Evaluation and Comprehensive Analysis
# =============================================================================

def evaluate_final_model(model, X_test, y_test, label_encoder):
    """Comprehensive evaluation with detailed metrics"""
    
    print(f"\nğŸ“Š FINAL MODEL EVALUATION")
    print("-" * 40)
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)
    
    # Calculate comprehensive metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision, recall, f1, support = precision_recall_fscore_support(y_test, y_pred, average='weighted')
    
    print(f"ğŸ¯ FINAL TEST RESULTS:")
    print(f"   Accuracy:  {accuracy:.6f} ({accuracy*100:.4f}%)")
    print(f"   Precision: {precision:.6f}")
    print(f"   Recall:    {recall:.6f}")
    print(f"   F1-Score:  {f1:.6f}")
    
    # Per-class metrics
    precision_per_class, recall_per_class, f1_per_class, support_per_class = precision_recall_fscore_support(
        y_test, y_pred, average=None
    )
    
    print(f"\nğŸ“‹ PER-CLASS METRICS:")
    print("-" * 25)
    for i, class_name in enumerate(label_encoder.classes_):
        class_mask = (y_test == i)
        if np.sum(class_mask) > 0:
            class_accuracy = accuracy_score(y_test[class_mask], y_pred[class_mask])
        else:
            class_accuracy = 0.0
        print(f"{class_name.capitalize():>8}: Acc={class_accuracy:.4f}, "
              f"Prec={precision_per_class[i]:.4f}, Rec={recall_per_class[i]:.4f}, "
              f"F1={f1_per_class[i]:.4f}, Support={support_per_class[i]}")
    
    return y_pred, y_pred_proba, accuracy

def plot_comprehensive_results(y_test, y_pred, y_pred_proba, label_encoder, accuracy):
    """Create comprehensive visualization of results"""
    
    print(f"\nğŸ“ˆ CREATING COMPREHENSIVE VISUALIZATIONS")
    print("-" * 45)
    
    fig = plt.figure(figsize=(20, 15))
    
    # 1. Confusion Matrix
    ax1 = plt.subplot(2, 3, 1)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=label_encoder.classes_, 
                yticklabels=label_encoder.classes_,
                cbar_kws={'label': 'Count'})
    plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    
    # 2. Normalized Confusion Matrix
    ax2 = plt.subplot(2, 3, 2)
    cm_normalized = confusion_matrix(y_test, y_pred, normalize='true')
    sns.heatmap(cm_normalized, annot=True, fmt='.3f', cmap='Reds', 
                xticklabels=label_encoder.classes_, 
                yticklabels=label_encoder.classes_,
                cbar_kws={'label': 'Proportion'})
    plt.title('Normalized Confusion Matrix', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    
    # 3. Per-Class Performance
    ax3 = plt.subplot(2, 3, 3)
    precision_per_class, recall_per_class, f1_per_class, _ = precision_recall_fscore_support(
        y_test, y_pred, average=None
    )
    
    x = np.arange(len(label_encoder.classes_))
    width = 0.25
    
    bars1 = plt.bar(x - width, precision_per_class, width, label='Precision', color='skyblue')
    bars2 = plt.bar(x, recall_per_class, width, label='Recall', color='lightgreen')
    bars3 = plt.bar(x + width, f1_per_class, width, label='F1-Score', color='lightcoral')
    
    plt.xlabel('Classes', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.title('Per-Class Performance Metrics', fontsize=14, fontweight='bold')
    plt.xticks(x, label_encoder.classes_, rotation=45)
    plt.legend()
    plt.ylim(0, 1.1)
    
    # Add value labels on bars
    for bars in [bars1, bars2, bars3]:
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{height:.3f}', ha='center', va='bottom', fontsize=8)
    
    # 4. Prediction Confidence Distribution
    ax4 = plt.subplot(2, 3, 4)
    max_proba = np.max(y_pred_proba, axis=1)
    correct_predictions = (y_test == y_pred)
    
    plt.hist(max_proba[correct_predictions], bins=20, alpha=0.7, label='Correct', color='green', density=True)
    plt.hist(max_proba[~correct_predictions], bins=20, alpha=0.7, label='Incorrect', color='red', density=True)
    plt.xlabel('Maximum Prediction Probability', fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.title('Prediction Confidence Distribution', fontsize=14, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 5. Overall Performance Summary
    ax5 = plt.subplot(2, 3, 5)
    overall_precision, overall_recall, overall_f1, _ = precision_recall_fscore_support(
        y_test, y_pred, average='weighted'
    )
    
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    values = [accuracy, overall_precision, overall_recall, overall_f1]
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
    
    bars = plt.bar(metrics, values, color=colors)
    plt.title('Overall Performance Metrics', fontsize=14, fontweight='bold')
    plt.ylabel('Score', fontsize=12)
    plt.ylim(0, 1.1)
    
    # Add value labels
    for bar, value in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.02,
                f'{value:.4f}', ha='center', va='bottom', fontweight='bold')
    
    # Add target line
    plt.axhline(y=0.9, color='red', linestyle='--', linewidth=2, label='Target (90%)')
    plt.legend()
    
    # 6. Classification Report Heatmap
    ax6 = plt.subplot(2, 3, 6)
    
    # Create classification report as dict
    from sklearn.metrics import classification_report
    report_dict = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
    
    # Extract metrics for heatmap
    classes = label_encoder.classes_
    metrics_matrix = []
    for class_name in classes:
        metrics_matrix.append([
            report_dict[class_name]['precision'],
            report_dict[class_name]['recall'],
            report_dict[class_name]['f1-score']
        ])
    
    metrics_matrix = np.array(metrics_matrix)
    
    sns.heatmap(metrics_matrix, annot=True, fmt='.3f', cmap='YlOrRd',
                xticklabels=['Precision', 'Recall', 'F1-Score'],
                yticklabels=classes, cbar_kws={'label': 'Score'})
    plt.title('Detailed Performance Heatmap', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed classification report
    print(f"\nğŸ“‹ DETAILED CLASSIFICATION REPORT:")
    print("=" * 60)
    print(classification_report(y_test, y_pred, target_names=label_encoder.classes_, digits=6))

# Evaluate final model
if X_raw is not None:
    print("ğŸ“Š Evaluating final model...")
    test_predictions, test_probabilities, test_accuracy = evaluate_final_model(
        best_lr_model, X_test_final, y_test, label_encoder
    )
    
    # Create comprehensive visualizations
    plot_comprehensive_results(y_test, test_predictions, test_probabilities, label_encoder, test_accuracy)

# =============================================================================
# CELL 6: Project Summary and Achievement Analysis
# =============================================================================

def print_final_summary():
    """Print comprehensive project summary and achievement analysis"""
    
    print("\n" + "="*80)
    print("ğŸ“ ENHANCED LOGISTIC REGRESSION PROJECT SUMMARY")
    print("="*80)
    
    print(f"ğŸ“Š Dataset Information:")
    print(f"   â€¢ Total samples: {len(X_raw):,} (after augmentation)")
    print(f"   â€¢ Original classes: {len(label_encoder.classes_)} ({', '.join(label_encoder.classes_)})")
    print(f"   â€¢ Image size: 224x224 (optimized for CNN)")
    print(f"   â€¢ Final features: {X_train_final.shape[1]:,} (CNN + traditional)")
    
    print(f"\nğŸ”§ Advanced Enhancements Applied:")
    print(f"   âœ… Comprehensive data augmentation (15+ transformations)")
    print(f"   âœ… Multi-CNN feature extraction (VGG16 + ResNet50 + InceptionV3)")
    print(f"   âœ… Traditional computer vision features (LBP, edges, moments, histograms)")
    print(f"   âœ… Advanced preprocessing (variance selection, scaling, feature selection, PCA)")
    print(f"   âœ… Extensive hyperparameter tuning ({len(compatible_params):,} combinations tested)")
    print(f"   âœ… Cross-validation and stratified splits")
    
    print(f"\nğŸ§  CNN Feature Extraction:")
    print(f"   â€¢ VGG16 features: {cnn_features[:, :512].shape[1]} dimensions")
    print(f"   â€¢ ResNet50 features: {cnn_features[:, 512:512+2048].shape[1]} dimensions") 
    print(f"   â€¢ InceptionV3 features: {cnn_features[:, 512+2048:].shape[1]} dimensions")
    print(f"   â€¢ Traditional features: {traditional_features.shape[1]} dimensions")
    print(f"   â€¢ Combined features: {X_combined.shape[1]:,} â†’ {X_train_final.shape[1]} (after preprocessing)")
    
    print(f"\nğŸ“‹ Best Logistic Regression Configuration:")
    for param, value in best_params.items():
        print(f"   â€¢ {param}: {value}")
    
    print(f"\nğŸ“ˆ Performance Results:")
    print(f"   â€¢ Validation Accuracy: {validation_score:.6f} ({validation_score*100:.4f}%)")
    print(f"   â€¢ Test Accuracy: {test_accuracy:.6f} ({test_accuracy*100:.4f}%)")
    
    # Calculate improvement from baseline
    baseline_accuracy = 0.80  # Your mentioned 80% baseline
    improvement = test_accuracy - baseline_accuracy
    improvement_percent = (improvement / baseline_accuracy) * 100
    
    print(f"\nğŸ“Š Performance Analysis:")
    print(f"   â€¢ Baseline (previous): {baseline_accuracy*100:.1f}%")
    print(f"   â€¢ Current result: {test_accuracy*100:.4f}%")
    print(f"   â€¢ Absolute improvement: +{improvement*100:.4f} percentage points")
    print(f"   â€¢ Relative improvement: +{improvement_percent:.2f}%")
    
    # Target achievement analysis
    target = 0.90
    if test_accuracy >= target:
        print(f"\nğŸ‰ ğŸ‰ ğŸ‰ TARGET ACHIEVED! ğŸ‰ ğŸ‰ ğŸ‰")
        print(f"âœ… Target: {target*100:.0f}% - ACHIEVED!")
        print(f"ğŸ† Final accuracy: {test_accuracy*100:.4f}%")
        print(f"ğŸš€ Exceeded target by: +{(test_accuracy-target)*100:.4f} percentage points")
        print(f"\nğŸŒŸ KEY SUCCESS FACTORS:")
        print(f"   1. CNN feature extraction provided rich representations")
        print(f"   2. Comprehensive data augmentation increased dataset size")
        print(f"   3. Advanced preprocessing improved feature quality")
        print(f"   4. Extensive hyperparameter tuning found optimal configuration")
        print(f"   5. Combined traditional + deep learning features")
    else:
        gap = target - test_accuracy
        print(f"\nğŸ“ˆ PROGRESS TOWARD TARGET:")
        print(f"   â€¢ Target: {target*100:.0f}%")
        print(f"   â€¢ Current: {test_accuracy*100:.4f}%")
        print(f"   â€¢ Remaining gap: {gap*100:.4f} percentage points")
        print(f"\nğŸ’¡ Recommendations for further improvement:")
        print(f"   â€¢ Fine-tune pre-trained CNN models")
        print(f"   â€¢ Add more sophisticated augmentation")
        print(f"   â€¢ Ensemble multiple models")
        print(f"   â€¢ Collect more training data")
        print(f"   â€¢ Try advanced regularization techniques")
    
    print(f"\nğŸ”¬ Technical Achievements:")
    print(f"   âœ… Successfully integrated CNN features with traditional CV features")
    print(f"   âœ… Implemented comprehensive data augmentation pipeline")
    print(f"   âœ… Applied advanced preprocessing and feature selection")
    print(f"   âœ… Performed extensive hyperparameter optimization")
    print(f"   âœ… Created detailed evaluation and visualization framework")
    
    print(f"\nğŸ’¾ Model Ready for Deployment:")
    print(f"   â€¢ Trained model: best_lr_model")
    print(f"   â€¢ Preprocessing pipeline: var_selector â†’ scaler â†’ feature_selector â†’ pca")
    print(f"   â€¢ Label encoder: label_encoder")
    print(f"   â€¢ Performance metrics: {test_accuracy*100:.4f}% accuracy")
    
    print("="*80)
    print("ğŸš€ ENHANCED LOGISTIC REGRESSION PROJECT COMPLETED SUCCESSFULLY!")
    print("="*80)

# Print comprehensive summary
if X_raw is not None:
    print_final_summary()
    
    # Save model and results
    import pickle
    
    model_package = {
        'model': best_lr_model,
        'variance_selector': var_selector,
        'scaler': scaler,
        'feature_selector': feature_selector,
        'pca': pca,
        'label_encoder': label_encoder,
        'best_params': best_params,
        'test_accuracy': test_accuracy,
        'validation_accuracy': validation_score
    }
    
    print(f"\nğŸ’¾ Model package ready for saving with pickle")
    print(f"ğŸ“Š All preprocessing steps included for inference pipeline")

print(f"\nğŸ¯ Enhanced Logistic Regression Pipeline Complete!")
print(f"ğŸ† Target: >90% Accuracy with CNN Features + Advanced ML")
print(f"âœ… Implementation: Complete with comprehensive evaluation")
```
